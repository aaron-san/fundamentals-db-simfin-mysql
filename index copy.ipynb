```python
DROP TABLE IF EXISTS us_shareprices_monthly;
-- When you drop the table, all indexes disappear automatically, so you must recreate them.

-- CREATE TABLE shareprices_monthly AS
-- SELECT *
-- FROM (
--     SELECT
--         Ticker,
--         SimFinId,
--         DATE_FORMAT(Date, '%Y-%m') AS Month,
--         Date,
--         Close,
--         AdjClose,
--         ROW_NUMBER() OVER (
--             PARTITION BY SimFinId, DATE_FORMAT(Date, '%Y-%m')
--             ORDER BY Date DESC
--         ) AS rn
--     FROM us_shareprices_daily_filtered
-- ) x
-- WHERE rn = 1;

SELECT * FROM us_shareprices_monthly LIMIT 10;

CREATE TABLE us_shareprices_monthly AS
SELECT
    Ticker,
    SimFinId,
    MAX(Date) AS Date,
    -- GROUP_CONCAT() → builds a list of Close prices
	-- ORDER BY Date DESC → newest price first
	-- SUBSTRING_INDEX(..., ',', 1) → pick the first price
    SUBSTRING_INDEX(
        GROUP_CONCAT(Close ORDER BY Date DESC),
        ',',
        1
    ) AS Close,
    SUBSTRING_INDEX(
        GROUP_CONCAT(AdjClose ORDER BY Date DESC),
        ',',
        1
    ) AS AdjClose
FROM us_shareprices_daily_filtered
GROUP BY
    Ticker,
    SimFinId,
    YEAR(Date),
    MONTH(Date);
    
    
CREATE INDEX idx_month_end_ticker_date
ON us_shareprices_monthly (SimFinId, Date);
```

## Metrics

### Miscellaneous Metrics
Next, miscellaneous metrics such as effective tax rate were computed and inserted into a new table, `metrics_misc`

```python
DROP TABLE IF EXISTS metrics_misc;

CREATE TABLE metrics_misc AS (
    WITH
        cte AS (
            SELECT
                isq.Ticker,
                isq.ReportDate,
                bsq.CashAndEquiv,
                CASE
                    WHEN isq.IncomeTaxExpenseBenefitNet >= 0 THEN NULL
                    WHEN isq.PretaxIncomeLoss <= 0 THEN NULL
                    ELSE -1 * isq.IncomeTaxExpenseBenefitNet / NULLIF(isq.PretaxIncomeLoss, 0)
                END AS EffectiveTaxRate,
                cfq.SharesBasic * (
                    SELECT pr.Close
                    FROM us_shareprices_daily pr
                    WHERE
                        pr.Ticker = isq.Ticker
                        AND pr.Date <= isq.ReportDate
                    ORDER BY pr.Date DESC
                    LIMIT 1
                ) AS MarketCap,
                bsq.ShortTermDebt + bsq.LongTermDebt AS TotalDebt,
                cfq.NetCashFromOperatingActivities + cfq.ChangeInFixedAssetsIntangibles AS FreeCashFlow
            FROM
                us_income_quarterly isq
                LEFT JOIN us_cashflow_quarterly cfq ON isq.Ticker = cfq.Ticker
                AND isq.ReportDate = cfq.ReportDate
                LEFT JOIN us_balance_quarterly bsq ON isq.Ticker = bsq.Ticker
                AND isq.ReportDate = bsq.ReportDate
        )
    SELECT cte.*, cte.MarketCap + cte.TotalDebt - cte.CashAndEquiv AS EnterpriseValue
    FROM cte
)
```

### Metrics Growth

```python
DROP TABLE IF EXISTS metrics_growth;

CREATE TABLE metrics_growth AS
SELECT
    isq.Ticker,
    isq.ReportDate,
    Revenue / NULLIF(LAG(Revenue, 4) OVER (
        PARTITION BY
            Ticker
        ORDER BY ReportDate
    ), 0) -1 AS RevenueGrowthYOY,
    CASE
        WHEN GrossProfit < 0
        OR LAG(GrossProfit, 4) OVER(PARTITION BY Ticker ORDER BY ReportDate) <= 0 THEN NULL
        ELSE GrossProfit / NULLIF(LAG(GrossProfit, 4) OVER (
            PARTITION BY
                Ticker
            ORDER BY ReportDate
        ), 0) -1
    END AS GrossProfitGrowthYOY,
    CASE
        WHEN OperatingIncomeLoss < 0
        OR LAG(OperatingIncomeLoss, 4) OVER(PARTITION BY Ticker ORDER BY ReportDate) <= 0 THEN NULL
        ELSE OperatingIncomeLoss / NULLIF(LAG(OperatingIncomeLoss, 4) OVER (
            PARTITION BY
                Ticker
            ORDER BY ReportDate
        ), 0) -1 
    END AS OperatingIncomeGrowthYOY,
    CASE
        WHEN NetIncome < 0
        OR LAG(NetIncome, 4) OVER(PARTITION BY Ticker ORDER BY ReportDate) <= 0 THEN NULL
        ELSE NetIncome / NULLIF(LAG(NetIncome, 4) OVER (
            PARTITION BY
                Ticker
            ORDER BY ReportDate
        ), 0) -1
    END AS NetIncomeGrowthYOY,
    TotalAssets / NULLIF(LAG(TotalAssets, 4) OVER (
        PARTITION BY
            Ticker
        ORDER BY ReportDate
    ), 0) -1 AS TotalAssetsGrowthYOY,
    TotalLiabilities / NULLIF(LAG(TotalLiabilities, 4) OVER (
        PARTITION BY
            Ticker
        ORDER BY ReportDate
    ), 0) -1 AS TotalLiabilitiesGrowthYOY,
    CashAndEquiv / NULLIF(LAG(CashAndEquiv, 4) OVER (
        PARTITION BY
            Ticker
        ORDER BY ReportDate
    ), 0) -1 AS CashAndEquivGrowthYOY,
    NetCashFromOperatingActivities / NULLIF(LAG(NetCashFromOperatingActivities, 4) OVER (
        PARTITION BY
            Ticker
        ORDER BY ReportDate
    ), 0) -1 AS NetCashFromOperatingActivitiesYOY,
    NetCashFromInvestingActivities / NULLIF(LAG(NetCashFromInvestingActivities, 4) OVER (
        PARTITION BY
            Ticker
        ORDER BY ReportDate
    ), 0) -1 AS NetCashFromInvestingActivitiesYOY,
    NetCashFromFinancingActivities / NULLIF(LAG(NetCashFromFinancingActivities, 4) OVER (
        PARTITION BY
            Ticker
        ORDER BY ReportDate
    ), 0) -1 AS NetCashFromFinancingActivitiesYOY 
FROM
    us_income_quarterly isq
    LEFT JOIN us_balance_quarterly bsq ON isq.Ticker = bsq.Ticker
    AND isq.ReportDate = bsq.ReportDate
    LEFT JOIN us_cashflow_quarterly cfq ON isq.Ticker = cfq.Ticker
    AND isq.ReportDate = cfq.ReportDate
```

### Profitability

Next, profitability metrics for each company were computed and a new table, `profitability_metrics` was created. The profitability metrics include operating profit margin, profit margin, 

```python
DROP TABLE IF EXISTS metrics_profitability;

CREATE TABLE metrics_profitability AS
SELECT
    isq.Ticker,
    isq.ReportDate,
    isq.GrossProfit / NULLIF(isq.Revenue, 0) AS GrossMargin,
    isq.OperatingIncomeLoss / NULLIF(isq.Revenue, 0) AS OperatingMargin,
    isq.NetIncome / NULLIF(isq.Revenue, 0) AS ProfitMargin,
    isq.NetIncome / NULLIF(
        LAG(bsq.TotalAssets, 1) OVER (
            PARTITION BY
                bsq.Ticker
            ORDER BY bsq.ReportDate
        ),
        0
    ) AS ROA,
    isq.NetIncome / NULLIF(
        LAG(bsq.TotalEquity, 1) OVER (
            PARTITION BY
                bsq.Ticker
            ORDER BY bsq.ReportDate
        ),
        0
    ) AS ROE,
    isq.OperatingIncomeLoss * (1 - misc.EffectiveTaxRate) / (
        NULLIF(bsq.TotalEquity, 0) + NULLIF(bsq.ShortTermDebt, 0) + NULLIF(bsq.LongTermDebt, 0) - NULLIF(bsq.CashAndEquiv, 0)
    ) AS ROIC,
    isq.OperatingIncomeLoss / NULLIF(
        LAG(misc.EnterpriseValue, 1) OVER (
            PARTITION BY
                misc.Ticker
            ORDER BY misc.ReportDate
        ),
        0
    ) AS EBITToEV
FROM
    us_income_quarterly isq
    LEFT JOIN us_cashflow_quarterly cfq ON isq.Ticker = cfq.Ticker
    AND isq.ReportDate = cfq.ReportDate
    LEFT JOIN us_balance_quarterly bsq ON isq.Ticker = bsq.Ticker
    AND isq.ReportDate = bsq.ReportDate
    LEFT JOIN metrics_misc misc ON isq.Ticker = misc.Ticker
    AND isq.ReportDate = misc.ReportDate;
```

```python
DROP TABLE IF EXISTS metrics_activity;

CREATE TABLE metrics_activity AS
SELECT
    isq.Ticker,
    isq.ReportDate,
    (
        SELECT pr.Date
        FROM us_shareprices_daily pr
        WHERE
            pr.Ticker = isq.Ticker
            AND pr.Date >= isq.ReportDate
        ORDER BY pr.Date ASC
        LIMIT 1
    ) AS PricesDate,
    (
        SELECT pr.Close
        FROM us_shareprices_daily pr
        WHERE
            pr.Ticker = isq.Ticker
            AND pr.Date >= isq.ReportDate
        ORDER BY pr.Date ASC
        LIMIT 1
    ) / NULLIF(isq.NetIncome, 0) AS PE,
    (
        SELECT pr.Close
        FROM us_shareprices_daily pr
        WHERE
            pr.Ticker = isq.Ticker
            AND pr.Date >= isq.ReportDate
        ORDER BY pr.Date ASC
        LIMIT 1
    ) / NULLIF(isq.Revenue, 0) AS PS,
    (
        SELECT pr.Close
        FROM us_shareprices_daily pr
        WHERE
            pr.Ticker = isq.Ticker
            AND pr.Date >= isq.ReportDate
        ORDER BY pr.Date ASC
        LIMIT 1
    ) / NULLIF(bsq.TotalEquity, 0) AS PB
FROM
    us_income_quarterly isq
    LEFT JOIN us_cashflow_quarterly cfq ON isq.Ticker = cfq.Ticker
    AND isq.ReportDate = cfq.ReportDate
    LEFT JOIN us_balance_quarterly bsq ON isq.Ticker = bsq.Ticker
    AND isq.ReportDate = bsq.ReportDate
    LEFT JOIN metrics_misc misc ON isq.Ticker = misc.Ticker
    AND isq.ReportDate = misc.ReportDate;
```

## Value Metrics

```python
DROP TABLE IF EXISTS metrics_value;

CREATE TABLE metrics_value AS
WITH
    ratios_price_date AS (
        SELECT isq.Ticker, isq.ReportDate, (
                SELECT pr.Date
                FROM us_shareprices_daily pr
                WHERE
                    pr.Ticker = isq.Ticker
                    AND pr.Date >= isq.ReportDate
                ORDER BY pr.Date ASC
                LIMIT 1
            ) AS PricesDate
        FROM
            us_shareprices_daily pr
            LEFT JOIN us_income_quarterly isq ON pr.Ticker = isq.Ticker
            AND isq.ReportDate = pr.Date
    )
SELECT isq.Ticker, isq.ReportDate, rpd.PricesDate, (
        SELECT pr.Close
        FROM us_shareprices_daily pr
        WHERE
            pr.Ticker = isq.Ticker
            AND pr.Date = rpd.PricesDate
    ) / 
    CASE 
        WHEN isq.NetIncome IS NULL OR isq.NetIncome < 0 THEN NULL
        ELSE (
            SELECT misc.MarketCap 
            FROM us_shareprices_daily pr
            WHERE
                pr.Ticker = isq.Ticker
                AND misc.Date = rpd.PricesDate
        ) / isq.NetIncome
    END AS PE, (
        SELECT misc.MarketCap
        FROM metrics_misc misc
        WHERE
            misc.Ticker = isq.Ticker
            AND misc.Date = rpd.PricesDate
    ) / NULLIF(isq.Revenue, 0) AS PS, (
        SELECT misc.MarketCap
        FROM metrics_misc misc
        WHERE
            misc.Ticker = isq.Ticker
            AND misc.Date = rpd.PricesDate
    ) / NULLIF(bsq.TotalEquity, 0) AS PB
FROM
    us_income_quarterly isq
    LEFT JOIN us_cashflow_quarterly cfq ON isq.Ticker = cfq.Ticker
    AND isq.ReportDate = cfq.ReportDate
    LEFT JOIN us_balance_quarterly bsq ON isq.Ticker = bsq.Ticker
    AND isq.ReportDate = bsq.ReportDate
    LEFT JOIN metrics_misc misc ON isq.Ticker = misc.Ticker
    AND isq.ReportDate = misc.ReportDate
    LEFT JOIN ratios_price_date rpd ON isq.Ticker = rpd.Ticker
    AND isq.ReportDate = rpd.ReportDate;
```

### Liquidity

Next, liquidity metrics for each company were computed and a new table, `liquidity_metrics` was created. The liquidity metrics include the current ratio, , profit margin, 

```python
-- Active: 1783290260690@@127.0.0.1@3306@fundamentals_data

DROP TABLE IF EXISTS metrics_liquidity;

CREATE TABLE metrics_liquidity AS (
WITH cte AS (
    SELECT
        bsq.Ticker,
        bsq.ReportDate,
        bsq.CashAndEquiv,
        bsq.TotalCurrentAssets,
        bsq.TotalCurrentLiabilities,
        bsq.Inventories,
        LAG(bsq.TotalCurrentAssets)
            OVER (PARTITION BY bsq.Ticker ORDER BY bsq.ReportDate) AS PrevCurrentAssets,
        LAG(bsq.TotalCurrentLiabilities)
            OVER (PARTITION BY bsq.Ticker ORDER BY bsq.ReportDate) AS PrevCurrentLiabilities
    FROM us_balance_quarterly bsq
)
SELECT
    cte.Ticker,
    cte.ReportDate,
    cte.TotalCurrentAssets,
    cte.TotalCurrentLiabilities,
    cte.CashAndEquiv / NULLIF(cte.TotalCurrentAssets, 0) AS CashRatio,
    (cte.TotalCurrentAssets - cte.Inventories) / NULLIF(cte.TotalCurrentAssets, 0) AS QuickRatio,
    (cte.TotalCurrentAssets / NULLIF(cte.TotalCurrentLiabilities, 0)) AS CurrentRatio,
    (cte.TotalCurrentAssets - cte.PrevCurrentAssets) / NULLIF(cte.PrevCurrentAssets, 0) AS CurrentAssetChg
FROM cte
WHERE cte.TotalCurrentLiabilities != 0
)
```
